# Effect Of Having A Partner On Monthly Bills

```python
sns.catplot(x="Partner", y="MonthlyCharges",kind="swarm", data=df)
plt.show()
```

```python
sns.catplot(x="Partner", y="MonthlyCharges", kind="boxen",
            data=df,aspect=2)
plt.show()
```

```python
sns.catplot(x="Partner", y="MonthlyCharges", kind="boxen",
            data=df,aspect=2,hue="gender")
plt.show()
```

```python
sns.catplot(x="Partner", y="MonthlyCharges", hue="gender", kind="bar", data=df)
plt.show()
```

Men Tend To Have Significantly Higher Monthly Bills When Having A Partner<br/>
Women Tend To Have Higher Monthly Bills When Having A Partner, But Not As Higher As Men<br/>

# Effect Of Having Dependents On Monthly Bills

```python
sns.catplot(x="Dependents", y="MonthlyCharges", kind="boxen",
            data=df,aspect=2)
plt.show()
```

Users Without Dependents Tend To Have Higher Monthly Bills<br/>
